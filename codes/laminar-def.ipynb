```python
import numpy as np
import matplotlib.pyplot as plt

# X-axis (flame progression direction)
x = np.linspace(0, 10, 500)

# Reactants: Exponentially decreasing
reactants = 1 - np.exp(-0.5 * (x - 2))
reactants[reactants < 0] = 0  # Prevent negative values

# Products: Complementary to reactants
products = 1 - reactants

# Temperature: S-curve shape
temperature = 1 / (1 + np.exp(-1.5 * (x - 5))) * 1200  # Temperature increases up to 1200K

# Reaction rate: Gaussian distribution shape
reaction_rate = np.exp(-0.5 * (x - 5)**2) * 3

# Plot setup
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot reactants, products, and reaction rate
ax1.plot(x, reactants, label="Reactants", color="blue", linewidth=2)
ax1.plot(x, products, label="Products", color="red", linewidth=2)
ax1.plot(x, reaction_rate, label="Reaction Rate", color="gray", linewidth=3, linestyle='--')

# Temperature plot (Secondary Y-axis)
ax2 = ax1.twinx()
ax2.plot(x, temperature, label="Temperature", color="orange", linewidth=2, linestyle=":")

# Highlight zones (Preheat Zone, Reaction Zone)
ax1.axvspan(2, 4, color='orange', alpha=0.2, label="Preheat Zone")
ax1.axvspan(4, 6, color='gray', alpha=0.2, label="Reaction Zone")

# Axis labeling
ax1.set_xlabel("Flame Position", fontsize=12)
ax1.set_ylabel("Concentration (mol/m) / Reaction Rate (1/s)", fontsize=12)
ax2.set_ylabel("Temperature (K)", fontsize=12)

# Legend
fig.legend(loc="upper right", bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)

# Graph styling
plt.title("Schematic Diagram of a Laminar Premixed Flame", fontsize=14)
ax1.grid()
plt.show()
```
