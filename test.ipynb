```python
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas=ct.Solution('gri30.yaml')
gas.TP = 300, 1e5
gas.set_equivalence_ratio(1.0, 'H2:1', 'O2:1, N2:3.76')
```

```python
premixed_flame=ct.FreeFlame(gas, width=0.02)
premixed_flame.inlet.X = gas.X
premixed_flame.inlet.T = gas.T
```

```python
premixed_flame.energy_enabled = True
premixed_flame.set_refine_criteria(ratio=2.0, slope=0.05, curve=0.05, prune=0.01)

premixed_flame.set_max_jac_age(10, 10)
premixed_flame.set_time_step(1e-8, [10, 20, 40, 80, 100, 100, 150])
premixed_flame.max_time_step_count = 500
```

```python
loglevel=0
premixed_flame.solve(loglevel, auto=True)
```

```python
plt.figure()
plt.plot(premixed_flame.grid, premixed_flame.T)
plt.ylabel('Temperature (K)')
plt.xlabel('Grid (m)')
plt.title('1D flame temperature profile')
plt.show()
```
